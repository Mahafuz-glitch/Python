```python
# course Management System
import sqlite3
d=sqlite3.connect('project.db')
c=d.cursor()
c.execute("create table pstud(sid text,sname text,saddr text)")
c.execute("create table pcrs(cid text,cname text,cfees text,cduration text)")
c.execute("create table pstatus(cid text,sid text,sname text,cname)")
d.commit()
print("Table created...")
```

```python
def getJoin(sid,sname):
    print("Select course id from list:")
    c.execute("select * from pcrs")
    res=c.fetchall()
    for r in res:
        print(r)
    cid=input("select your course id:")
    c.execute("select cname from pcrs where cid=?",(cid,))
    res=c.fetchall()
    if(res):
        cname=res[0][0]
        print("Your course name :",cname)
        print("Your course id :",cid)
        print("Your registration id :",sid)
        print("Your name :",sname)
        c.execute("insert into pstatus values(?,?,?,?)",(cid,sid,sname,cname,))
        d.commit()
        print("Congratulation...project complete")
    else:
        print("Wrong course id")
    
def viewcourse():
    c.execute("select * from pcrs")
    res=c.fetchall()
    for r in res:
        print(r)
def regis():
    sid=input("Enter id :")
    sname=input("Enter name :")
    saddr=input("Enter address :")
    c.execute("insert into pstud values(?,?,?)",(sid,sname,saddr,))
    d.commit()
    print("registration success")
def admission():
    sid=input("please enter your id :")
    c.execute("select sname from pstud where sid=?",(sid,))
    res=c.fetchall()
    if(res):
        sname=res[0][0]
        print("Hello",sname,"press enter...")
        getJoin(sid,sname)
        
        
    else:
        print("No record found")
def activity():
    print("press 1 for course details")
    print("press 2 for student details")
    print("press 3 for course modification")
    print("press 4 to admission status")
    chs=input("enter choice :")
    if(chs=='1'):
        c.execute("select * from pcrs")
        rs=c.fetchall()
        for i in rs:
            print(i)
    elif(chs=='2'):
        c.execute("select * from pstud")
        rs=c.fetchall()
        for i in rs:
            print(i)
    elif(chs=='3'):
        print("press 1 to insert course")
        print("press 2 to delete course")
        print("press 3 to update course")
        cs=input("Please enter :")
        if(cs=='1'):
            cid=input("Enter new course id:")
            cname=input("Enter new course name :")
            cfees=input("Enter course fees :")
            cduration=input("Enter duration :")
            c.execute("insert into pcrs values(?,?,?,?)",(cid,cname,cfees,cduration))
            d.commit()
            print("course saved")
        elif(cs=='2'):
            cid=input("Enter course id :")
            c.execute("select * from pcrs where cid=?",(cid,))
            rs=c.fetchall()
            if(rs):
                c.execute("delete from pcrs where cid=?",(cid))
            else:
                print("Wrong ID")
        elif(cs=='3'):
            cid=input("Enter courseid for update:")
            c.execute("select cfees,cduration from pcrs where cid=?",(cid,))
            rs1=c.fetchall()
            print("Old structure")
            for i in rs1:
                print(i)
            cfees=input("Enter new fees :")
            cduration=input("Enter new duration :")
            c.execute("update pcrs set cfees=?,cduration=? where cid=?",(cfees,cduration,cid))
            d.commit()
            print("Course updated....")
    elif(chs=='4'):
        c.execute("select * from pstatus")
        res=c.fetchall()
        for r in res:
            print(r)
            
    else:
        print("Wrong choice")
    
def admin():
    aid=input("Enter admin ID :")
    if(aid=="admin"):
        activity()
    else:
        print("Wrong ID")
def stud():
    print("press 1 for course information")
    print("press 2 for registration")
    print("press 3 for admission")
    print("press 4 for main menu")
    chs=input("select option:")
    if(chs=='1'):
        viewcourse()
    elif(chs=='2'):
        regis()
    elif(chs=='3'):
        admission()
    
    else:
        print("Wrong choice")
    
    
print("************************")
print("Course Management System")
print("************************")
ans='y'
while(ans=='y'):
    print("press 1 for admin login")
    print("press 2 for student login")
    chs=input("Please enter :")
    if(chs=='1'):
        admin()
    elif(chs=='2'):
        stud()
    else:
        print("Wrong choice")
    ans=input("Want to continue(y/n):")
    if(ans=='y'):
        print("Continue")
print("Thanks...")
```
