```python
df['Customer Complaint']=df['Customer Complaint'].apply(lambda x : str(x).lower())
```

```python
df['Customer Complaint'].value_counts()[:10].plot.bar()
```

```python
df['Status'].unique()
```

```python
df['new_status']=['Open' if st=="Open" or st=="Pending"  else "Closed"    for st in df['Status']]
```

```python
df.head()
```

```python
df["new_status"].unique()
```

```python
state_complain=df.groupby(['State','new_status']).size().unstack().fillna(0)
```

```python
state_complain
```

```python
state_complain.plot.bar(stacked=True,figsize=(10,10))
```

```python
#Which state has the maximum complaints

df.groupby(['State']).size().sort_values(ascending=False)[:5]
```

```python
#Which state has the highest percentage of unresolved complaints

unresolved_data=df.groupby(['State','new_status']).size().unstack().fillna(0).sort_values(by='Open',ascending=False)
```

```python
unresolved_data['unresolved_cmp_prct']=unresolved_data['Open']/unresolved_data['Open'].sum()*100
```

```python
unresolved_data
```

```python


resolved_data=df.groupby(['Received Via','new_status']).size().unstack()
```

```python
resolved_data['resloved']=resolved_data['Closed']/resolved_data['Closed'].sum()*100
```
