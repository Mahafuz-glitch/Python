```python
#Barplot

names=["Sam","Amit","Deepak"]
age=[87,56,27]
```

```python
plt.bar(names,age)
plt.title("Bar Plot")
plt.xlabel("Names")
plt.ylabel("Age")
plt.show()
```

```python
plt.bar(names,age,color=['red','green','yellow'])
plt.title("Bar Plot")
plt.xlabel("Names")
plt.ylabel("Age")
plt.show()
```

```python
plt.barh(names,age,color=['red','green','yellow'])
plt.title("Bar Plot")
plt.xlabel("Age")
plt.ylabel("Name")
plt.show()
```

```python
plt.bar(names,age,color=['red','green','yellow'])
plt.text(0,age[0],age[0])
plt.text(1,age[1],age[1])
plt.text(2,age[2],age[2])
plt.title("Bar Plot")
plt.xlabel("Names")
plt.ylabel("Age")
plt.show()
```
